## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as st


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
#see sample of mouse data
mouse_metadata.head()
```

```python
#see sample of study data
study_results.head()
```

```python
#count unique mice id in original data to compare clean df to
mouse_count = len(study_results["Mouse ID"].value_counts())
mouse_count
```

```python
#create clean data frame without timepoint duplicates
clean_study = study_results.drop_duplicates(subset=["Mouse ID", "Timepoint"])
#check that we didn't lose any mice
clean_mouse_check = len(clean_study["Mouse ID"].value_counts())
clean_mouse_check
```

```python
# Combine the data into a single dataset
merged_df = pd.merge(mouse_metadata, clean_study, how="outer", on="Mouse ID")
# Display the data table for preview
merged_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
merged_mouse_check = len(merged_df["Mouse ID"].value_counts())
merged_mouse_check
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
drug_group = merged_df.groupby("Drug Regimen")
mean = drug_group["Tumor Volume (mm3)"].mean()
median = drug_group["Tumor Volume (mm3)"].median()
variance = drug_group["Tumor Volume (mm3)"].var()
std = drug_group["Tumor Volume (mm3)"].std()
sem = drug_group["Tumor Volume (mm3)"].sem()
summary_df = pd.DataFrame({"mean": mean, "median": median, "variance": variance, "standard_deviation": std, "sem": sem})
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
drug_group["Tumor Volume (mm3)"].agg(["mean", "median","var", "std", "sem"])
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
bar_plot = drug_group["Drug Regimen"].count().plot.bar()
bar_plot.set_xlabel("Drug Regimen")
bar_plot.set_ylabel("Count of Data Points")
bar_plot.set_title("Drug Regimen vs. # of Data Points")
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.
#set x axis and tick locations
x_axis = np.arange(len(drug_group))
tick_locations = [x for x in x_axis]
plt.bar(x_axis, drug_group["Drug Regimen"].count())
plt.xticks(tick_locations,list(summary_df.index), rotation = "vertical")
plt.title("Drug Regimen vs. # of Data Points")
plt.xlabel("Drug Regimen")
plt.ylabel("Count of Data Points")
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
gender_group = merged_df.groupby("Sex")
gender_counts = pd.DataFrame(gender_group["Sex"].count())
gender_counts.plot(kind="pie",y="Sex", title = "Male vs. Female", autopct='%1.1f%%', legend=False, startangle=120)
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot

genders=list(gender_counts.index.values)
pyplot_gender_count = gender_counts["Sex"]
plt.pie(pyplot_gender_count, labels = genders, autopct="%1.1f%%", startangle=120)
plt.title("Male vs. Female")
```
